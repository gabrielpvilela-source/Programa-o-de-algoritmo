```python
# Escreva um programa que o usúario digita uma palavra e o programa
# verifica se a palavra é palíndroma e imprime a palavra

# Entrada
invertida = ""
palavra = input("Digite uma palavra: ")
# Processamento
for i in (palavra):
    invertida = i + invertida
print (invertida)
  # Saída
if invertida == palavra:
  print ("A palavra é palíndroma")
else:
  print ("A palavra não é palíndroma")
```

```python
# Escreva um programa qure o usúario
# Digita um número e o programa imprime os números pares

número = int(input("Digite um número: "))


if número % 2 == 0:
  print ("O número é par")
else:
  print ("O número é ímpar")

for i in range (número):
  if i % 2 == 0:
    print (i)
```
