```python
idade = int(input("Digite a idade da pessoa: "))

if idade <= 17:
    print("Menor de idade")
elif 18 <= idade <= 59:
    print("Adulto")
else:
    print("Idoso")

valor_compra = float(input("Digite o valor da compra: "))

desconto = 0
if valor_compra <= 100:
    desconto = 0
elif 101 <= valor_compra <= 500:
    desconto = 0.10 # 10%
else:
    desconto = 0.20 # 20%

valor_final = valor_compra * (1 - desconto)
print(f"O valor final da compra com {desconto*100:.0f}% de desconto é R${valor_final:.2f}")



lado1 = float(input("Digite o valor do primeiro lado: "))
lado2 = float(input("Digite o valor do segundo lado: "))
lado3 = float(input("Digite o valor do terceiro lado: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("É um triângulo válido.")
else:
    print("Não é um triângulo válido.")


energia = int(input("Digite a energia do personagem (de 0 a 100): "))

if energia == 0:
    print("Morto")
elif 5 <= energia <= 30:
    print("Crítico")
elif 31 <= energia <= 70:
    print("Normal")
else:
    print("Saudável")



media_final = float(input("Digite a média final do aluno: "))

if media_final >= 7:
    print("Aprovado")
elif 5 <= media_final <= 6.9:
    print("Recuperação")
else:
    print("Reprovado")

salario = float(input("Digite o salário do funcionário: "))
pontuacao = int(input("Digite a pontuação de desempenho (0 a 10): "))

bonus_percentual = 0
if pontuacao < 5:
    bonus_percentual = 0
elif 5 <= pontuacao <= 7:
    bonus_percentual = 0.05 # 5%
elif 8 <= pontuacao <= 9:
    bonus_percentual = 0.10 # 10%
else:
    bonus_percentual = 0.15 # 15%

valor_bonus = salario * bonus_percentual
salario_final = salario + valor_bonus

print(f"Valor do bônus: R${valor_bonus:.2f}")
print(f"Salário final com bônus: R${salario_final:.2f}")
```
