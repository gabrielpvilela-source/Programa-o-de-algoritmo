```python
#  Operadores aritiméticos
#  Fazer um progerma que calcula operações com 02 números


# Entrda


a = int (input("digite um número: "))
b = int (input("digite um número: "))


# Processamento

soma = a + b
multiplicacao = a * b
divisao = a/b
subtracao = a-b
potencia = a ** b
divisao_inteira = a // b
resto = a % b

# Saída
print ("")
print("---------------------------------------------------OPERADORES ARITIMÉTICOS-------------------------------------------------")
print ("")
print ("-----------------------------------------------------SOMA-----------------------------------------------------")
print ("")
print (f"O resulado da SOMA de {a} MAIS {b} é igual a: {soma}")
print ("")
print ("------------------------------------------MULTIPLICACAO---------------------------------------------------------")
print (f"O resultado da MULTIPLICACAO de {a} por {b} é igual a: {multiplicacao}")
print ("")
print ("-------------------------------------------DIVISAO-------------------------------------------------------------")
print (f" O resultado de DIVISAO de {a} por {b} é igual a: {divisao}")
print ("")
print ("---------------------------------------------SUBTRACAO-----------------------------------------------------------")
print (f"O resultado da SUBTRACAO de {a} por {b} é igual a: {subtracao}")
print ("")
print ("-------------------------------------------POTENCIA----------------")
print (f"O resultado da POTENCIA de {a} por {b} é igual a: {potencia}")
print ("")
print("-------------------------------------DIVISAO INTEIRA----------------------------------")
print (f"O resultado da DIVISAO INTEIRA de {a} por {b} é igual: {divisao_inteira}")
print ("")
print ("-------------------------------------------------RESTO----------------------------------------")
print (f"O resultado do RESTO de {a} por {b} é igual a: {resto}")
print ("")
print ("------------------------------------------------------------------------------------------------------------------------")
```
