```python
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))
nota1 = float(input("Digite sua nota1: ").replace(',', '.'))
nota2 = float(input("Digite sua nota2: ").replace(',', '.'))
nota3 = float(input("Digite sua nota3: ").replace(',', '.'))
nota4 = float(input("Digite sua nota4: ").replace(',', '.'))

# Processamento

media = (nota1 + nota2 + nota3 + nota4) / 4

# Saída

print (f"olá {nome}, seja bem vindo!")
print (f"você tem {idade} anos")
print (f"sua média é igual {media}")
```
