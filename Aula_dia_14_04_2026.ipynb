```python
2
# Escreva um programa Solicite dois números e mostre qual é o maior (ou se são iguais).

# Entrada

Número1=int(input("Digite o primeiro número: "))
Número2=int(input("Digite o segundo número: "))

# Processamento/saída

if Número1 > Número2:
    print("O primeiro número é maior que o segundo.")
elif Número1 < Número2:
    print("O segundo número é maior que o primeiro.")
else:
    print("Os números são iguais.")
```

```python
3
# Escreva um programa Peça a nota de um aluno e informe:
# Aprovado (nota >= 7)
# Recuperação (5 <= nota < 7)
# Reprovado (nota < 5)

# Entrada
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")
elif nota >= 5 and nota < 7:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
4
# Escreva um programa Classificação de idade
# Receba a idade e classifique:
# Criança (0-12)
# Adolescente (13-17)
# Adulto (18-59)
# Idoso (60+)

# Entrada
Idade = int(input("Digite a idade: "))

# Processamento/saída

if Idade >= 0 and Idade <= 12:
    print("Criança")
elif Idade >= 13 and Idade <= 17:
    print("Adolescente")
elif Idade >= 18 and Idade <= 59:
    print("Adulto")
elif Idade >= 60:
    print("Idoso")
```

```python
7
# Escreva um programa Calculadora simples
# Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado

# Entrada
Número1=int(input("Digite o primeiro número: "))
Número2=int(input("Digite o segundo número: "))
Operação=input("Digite a operação (+, -, *, /): ")

# Processamento/saída

if Operação == '+':
    print("O resultado é:", Número1 + Número2)
elif Operação == '-':
    print("O resultado é:", Número1 - Número2)
elif Operação == '*':
    print("O resultado é:", Número1 * Número2)
elif Operação == '/':
    if Número2 != 0:
            print("O resultado é:", Número1 / Número2)
    else:
        print("Erro: Divisão por zero não é permitida.")
else:
    print("Operação inválida. Use +, -, * ou /.")
```

```python
8
# Escreva um programa Dia da semana
# Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

# Entrada
Número=int(input("Digite um número de 1 a 7: "))

# Processamento/saída

if Número == 1:
    print("Domingo")
elif Número == 2:
    print("Segunda-feira")
elif Número == 3:
    print("Terça-feira")
elif Número == 4:
    print("Quarta-feira")
elif Número == 5:
    print("Quinta-feira")
elif Número == 6:
    print("Sexta-feira")
elif Número == 7:
    print("Sábado")
else:
    print("Número inválido. Digite um número entre 1 e 7.")
```

```python
9
# Escreva um programa Solicite um ano e informe se ele é bissexto ou não.

# Entrada

Ano = int(input("Digite um ano: "))

# Processamento/saída

if (Ano % 4 == 0 and Ano % 100 != 0) or (Ano % 400 == 0):
    print(f"O ano {Ano} é bissexto.")
else:
    print(f"O ano {Ano} não é bissexto.")
```

```python
10
# Escreva um programa Peça o valor de uma compra:
# ≥ 100: 10% de desconto
# ≥ 50 e < 100: 5% de desconto
# < 50: sem desconto

# Entrada

valor_compra = float(input("Digite o valor da compra: "))

# Processamento/saída

if valor_compra >= 100:
    percentual_desconto = 0.10
elif valor_compra >= 50:
    percentual_desconto = 0.05
else:
    percentual_desconto = 0

valor_final = valor_compra * (1 - percentual_desconto)
print(f"Valor da compra: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {percentual_desconto * 100:.0f}%")
print(f"Valor final com desconto: R$ {valor_final:.2f}")
```
