```python
# Estrutura de decições


# Escrever um programa que recebe o nome ea idade do usuárioe verifica se é criança, adolescente e adulto


# >= 12 anos é criança
# =< 17 é adolecscente
# Adulto


# Entrada

nome = str (input ("Digite seu nome: "))
idade = int (input ("Digite sua idade: "))

# Processamento / Saída

if idade <= 12:
  print (f"{nome}, você tem {idade} anos de idade, você é criança.")
elif idade <= 18:
  print (f"{nome}, você tem {idade} anos de idade, você é adolescente.")
else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

```python
# Em um jogo, um personagem possui uma variável chamada energia. O programa deve:
# Se a energia for maior que 70 - Forte, se energia estiver entre 30 e 70 - Cansado,
# Se a energia estiver menor que 30 - Fraco.

# Entrada
Energia = int(input("digite a energia do seu personagem: "))

# Processamento / Saída
if Energia < 30:
  print(f" Seu personagem tem energia igual a {Energia}. Está fraco.")
elif Energia <= 70:
  print(f" Seu personagem tem energia igual a {Energia}. Está cansado.")
else:
  print(f" Seu personagem tem energia igual a {Energia}. Está forte.")
```

```python
# Escrever um programa que classifica os tipos de triângulos e valida se é um triângulo.

# Validar se os lados digitados podem formar um triângulo
# Lado1 + Lado2 > Lado3 and Lado3 + Lado1 > Lado2 and Lado2 + Lado3 > Lado1

# Tipos de triângulo
# Equilátero - todos os lados iguais - ==
# Escaleno - todos os lados diferentes- !=
# Isósceles - dois lados iguais

# Entrada

Lado1 = float(input("Digite o valor do primeiro lado: "))
Lado2 = float(input("Digite o valor do segundo lado: "))
Lado3 = float(input("Digite o valor do terceiro lado: "))

# Processamento / Saída
if (Lado1 + Lado2 > Lado3) and (Lado3 + Lado1 > Lado2) and (Lado2 + Lado3 > Lado1):
  print("Seu triângulo é válido.")
  if Lado1 == Lado2 == Lado3:
    print("Seu triângulo é equilátero.")
  elif Lado1 != Lado2 and Lado2 != Lado3 and Lado1 != Lado3:
    print("Seu triângulo é escaleno.")
  else: # Se é válido e não é equilátero nem escaleno, então é isósceles.
    print("Seu triângulo é isósceles.")
else:
  print("Seu triângulo não é válido.")
```
